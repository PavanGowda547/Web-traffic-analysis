```python
#Importing Required Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data 
data=pd.read_csv('website_wata.csv')
```

```python
#Inspect the data 
print("Displaying Basic information : ")
print(data.info())
```

```python
#Shape of the dataset
print("Shape of the Dataset (row, columns) : ",data.shape)
```

```python
print("\n First 5 rows :")
print(data.head())
```

```python
#Finding Duplicate Values
if data.duplicated().any():
    print(f"There are as many as {data.duplicated().sum()} duplicate data.")
else:
    print("There are no duplicate data.")
```

```python
traffic_count = data.groupby('Traffic Source').size().reset_index(name = 'count')
traffic_count
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(data = traffic_count, x = 'Traffic Source', y = 'count', color = 'skyblue' )
ax.bar_label(ax.containers[0], fontsize = 10)
plt.title('Traffic Source Distribution')
plt.xlabel('Traffic Source')
plt.ylabel('Count')
```

```python
#Plotting a Histogram with kde
sns.histplot(data['Page Views'], bins=15, kde=True, edgecolor='black', alpha=0.7)
plt.title('Distribution of Page Views', fontsize=16)
plt.xlabel('Page Views', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
session_duration = round(data.groupby('Traffic Source')['Session Duration'].mean().reset_index(name = 'avg_session_duration'),2)
session_duration
```

```python
plt.figure(figsize = (10, 6))
ax = sns.barplot(data = session_duration, x = 'Traffic Source', y = 'avg_session_duration')
ax.bar_label(ax.containers[0], fontsize = 10)
plt.title('Avg Session Duration')
plt.xlabel('Traffic Source')
plt.ylabel = ('Average duration (in secs)')
```
