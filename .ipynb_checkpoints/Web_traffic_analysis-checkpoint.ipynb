```python
#Importing Required Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sts
```

```python
#Load the data 
data=pd.read_csv('website_wata.csv')
```

```python
#Inspect the data 
print("Displaying Basic information : ")
print(data.info())
```

```python
#Inspect the data 
print("Displaying Object information : ")
print(data.dtypes)
```

```python
#Shape of the dataset
print("Shape of the Dataset (row, columns) : ",data.shape)
```

```python
print("\n First 5 rows :")
data.head()
```

```python
#Finding Duplicate Values
if data.duplicated().any():
    print(f"There are as many as {data.duplicated().sum()} duplicate data.")
else:
    print("There are no duplicate data.")
```

```python
#Finding null values
print(data.isnull().sum())
```

```python
traffic_count = data.groupby('Traffic Source').size().reset_index(name = 'count')
plt.figure(figsize=(10, 5))
ax = sns.barplot(data = traffic_count, x = 'Traffic Source', y = 'count', color = 'skyblue' )
ax.bar_label(ax.containers[0], fontsize = 10)
plt.title('Traffic Source Distribution')
plt.xlabel('Traffic Source')
plt.show()
```

```python
#Plotting a Histogram with kde of Page views
sns.histplot(data['Page Views'], bins=15, kde=True, edgecolor='black', alpha=0.7)
plt.title('Distribution of Page Views', fontsize=16)
plt.xlabel('Page Views', fontsize=14)
plt.show()
```

```python
#Visualizations of Bounce rate by Traffic source
plt.figure(figsize=(10, 6))
sns.boxplot(x="Traffic Source", y="Bounce Rate", data=data)
plt.title("Bounce Rate by Traffic Source", fontsize=16)
plt.xlabel("Traffic Source", fontsize=12)
plt.show()
```

```python
#Visualization of Average Session Duration 
session_duration = round(data.groupby('Traffic Source')['Session Duration'].mean().reset_index(name = 'avg_session_duration'),2)
plt.figure(figsize = (10, 6))
ax = sns.barplot(data = session_duration, x = 'Traffic Source', y = 'avg_session_duration')
ax.bar_label(ax.containers[0], fontsize = 10)
plt.title('Avg Session Duration')
plt.xlabel('Traffic Source')
plt.ylabel = ('Average duration (in secs)')
```

```python
corr = data.drop("Traffic Source", axis=1).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title('Correlation Heatmap of Website Metrics')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Traffic Source', y='Time on Page', data=data)
plt.title('Time on Page Distribution by Traffic Source')
plt.show()
```

```python
# T-Test for Session Duration between 'Organic' and 'Paid' Traffic Source
organic_sessions = data[data['Traffic Source'] == 'Organic']['Session Duration']
paid_sessions = data[data['Traffic Source'] == 'Paid']['Session Duration']

t_stat, p_value = sts.ttest_ind(organic_sessions, paid_sessions)
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.3f}")

# Interpret the result
if p_value < 0.05:
    print("There is a significant difference in Session Duration between Organic and Paid traffic sources.")
else:
    print("There is no significant difference in Session Duration between Organic and Paid traffic sources.")
```
